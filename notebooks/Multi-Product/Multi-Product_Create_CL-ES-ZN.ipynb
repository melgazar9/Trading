### NOTE - need to create full CL, ES and ZN dataframes (using \\<PRODUCT\>\_concat_new_data.py), then run the code below

```python
import pandas as pd
import os
```

```python
df_CL = pd.DataFrame()
filepath = '/media/melgazar9/95d0b8d6-6582-4af9-8c3e-d9497417f0d1/Trading/Data/CL/CL_TD_historical_data_DFs/CL_1min/'
for f in os.listdir(filepath):
    df_CL = pd.concat([df_CL, pd.read_csv(filepath + f)], axis=0, sort=False)
    df_CL.drop_duplicates(inplace=True)

df_CL = df_CL.drop_duplicates().set_index('Datetime')
df_CL.index = pd.to_datetime(df_CL.index)
df_CL = df_CL.add_prefix('CL_')
df_CL.sort_index(inplace=True)
print(df_CL.shape)
df_CL.head(2)
```

```python
df_CL.tail(2)
```

```python
df_ES = pd.DataFrame()
filepath = '/media/melgazar9/95d0b8d6-6582-4af9-8c3e-d9497417f0d1/Trading/Data/ES/ES_TD_historical_data_DFs/ES_1min/'
for f in os.listdir(filepath):
    df_ES = pd.concat([df_ES, pd.read_csv(filepath + f)], axis=0, sort=False)
    df_ES.drop_duplicates(inplace=True)
    
df_ES = df_ES.drop_duplicates().set_index('Datetime').sort_index()
df_ES.index = pd.to_datetime(df_ES.index)
df_ES = df_ES.add_prefix('ES_')
df_ES.sort_index(inplace=True)
print(df_ES.shape)
df_ES.head(2)
```

```python
df_ZN = pd.DataFrame()
filepath = '/media/melgazar9/95d0b8d6-6582-4af9-8c3e-d9497417f0d1/Trading/Data/ZN/ZN_TD_historical_data_DFs/ZN_1min/'
for f in os.listdir(filepath):
    df_ZN = pd.concat([df_ZN, pd.read_csv(filepath + f)], axis=0, sort=False)
    df_ZN.drop_duplicates(inplace=True)
    
df_ZN = df_ZN.drop_duplicates().set_index('Datetime').sort_index()
df_ZN.index = pd.to_datetime(df_ZN.index)
df_ZN = df_ZN.add_prefix('ZN_')
df_ZN.sort_index(inplace=True)
print(df_ZN.shape)
df_ZN.head(2)
```

```python
df_CL.shape, df_ES.shape, df_ZN.shape
```

```python
df = pd.merge(df_CL, df_ES, left_index=True, right_index=True) # Inner join
df = pd.merge(df, df_ZN, left_index=True, right_index=True) # Inner join
```

```python
df.drop_duplicates(inplace=True)
```
